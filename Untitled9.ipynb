```python
import pandas as pd
import csv
import numpy as py
```

```python
df=pd.read_csv("./data.csv")
```

```python
listcol=[]
for col in df.columns:
  if df[col].sum()<10:
    listcol.append(col)
```

```python
df.drop(columns=listcol,axis=1,inplace=True)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ax = plt.subplots()
sns.countplot(x='target', data=df, ax=ax)
ax.set_ylim(0, 200)
ax.set_title('Target Distribution', weight='bold')
plt.show()
```
